### Search KEGG 2.0 MINE for Masses in Sauer Test Data File

Now, negative

```python
ms_params = {
    'adducts': ["[M-H]-", "[M+F]-"],
    'tolerance': 2,  # mDa
    'ppm': False,
    'logp': None,
    'halogens': True,
    'verbose': False,
    'charge': '-',
    'models': []
}
```

```python
ids = mb_df.index
masses = [str(val) for val in mb_df['m/z'].values]
len(masses)
```

```python
hits_dict = {}
for cpd_id, mass in zip(ids, masses):
    print(cpd_id, mass)
    hits_dict[cpd_id] = ms_adduct_search(db=db, core_db=core_db, keggdb=None, text=mass, text_type='form', ms_params=ms_params)
```

### Calculate Statistics from Search Output

```python
n_annotated = 0
n_unannotated_annotated = 0
num_hits = []

for (cpd_id, hits), annotation in zip(hits_dict.items(), list(mb_df.Annotation.values)):
    if len(hits) > 0:
        n_annotated += 1
        if not isinstance(annotation, str) and np.isnan(annotation):
            n_unannotated_annotated += 1
    num_hits.append(len(hits))
```
